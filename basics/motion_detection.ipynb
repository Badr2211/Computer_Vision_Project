```python
import cv2 as cv
import numpy as np
from time import sleep

largura_min = 80
altura_min = 80
offset = 6
pos_linha = 550
delay = 60

detec = []
carros = 0

def pega_centro(x, y, w, h):
    x1 = int(w / 2)
    y1 = int(h / 2)
    cx = x + x1
    cy = y + y1
    return cx, cy  # Fixed the typo here

cap = cv.VideoCapture('highway.mp4')
subtracao = cv.bgsegm.createBackgroundSubtractorMOG(history=12)

while cap.isOpened():
    _, frame1 = cap.read()
    if _ is not True:
        break

    tempo = float(1 / delay)
    sleep(tempo)

    grey = cv.cvtColor(frame1, cv.COLOR_BGR2GRAY)
    blur = cv.GaussianBlur(grey, (3, 3), 5)
    
    img_sub = subtracao.apply(blur)
    dilate = cv.dilate(img_sub, np.ones((9, 9)))
    #kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5, 5))
    #dilatada = cv.morphologyEx(dilate, cv.MORPH_ELLIPSE, kernel)
    cv2.imshow('dilated',dilate)
    contorno, h = cv.findContours(dilate, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

    cv.line(frame1, (25, pos_linha), (1200, pos_linha), (176, 130, 39), 2)
    cv.line(frame1, (25,altura_min ), (1200, largura_min), (0, 0, 39), 2)

    for i, c in enumerate(contorno):
        x, y, w, h = cv.boundingRect(c)
        validar_contorno = (w >= largura_min) and (h >= altura_min)

        #if not validar_contorno:
            #continue
        
        cv.rectangle(frame1, (x, y), (x + w, y + h), (0, 255, 0), 2)
        centro = pega_centro(x, y, w, h)
        detec.append(centro)

    for x, y in detec.copy():  # Use copy to iterate over a copy of the list
        if pos_linha - offset < y < pos_linha + offset:
            carros += 1
            cv.line(frame1, (25, pos_linha), (1200, pos_linha), (0, 0, 255), 2)
            detec.remove((x, y))

    cv.putText(frame1, "VEHICLE COUNT : " + str(carros), (320, 70),
               cv.FONT_HERSHEY_COMPLEX, 2, (0, 0, 255), 4)

    cv.imshow("Vehicle Detection", frame1)

    if cv.waitKey(1) == 27:
        break

cv.destroyAllWindows()
cap.release()
```
