```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def create_blank_image(width, height):
    return np.zeros([width, height, 3], np.uint8)

def draw_line(image, start_point, end_point, color, thickness):
    return cv2.line(image, start_point, end_point, color, thickness)

def draw_arrowed_line(image, start_point, end_point, color, thickness):
    return cv2.arrowedLine(image, start_point, end_point, color, thickness)

def draw_rectangle(image, top_left, bottom_right, color, thickness):
    return cv2.rectangle(image, top_left, bottom_right, color, thickness)

def draw_circle(image, center, radius, color, thickness):
    return cv2.circle(image, center, radius, color, thickness)

def draw_text(image, text, position, font, font_size, color, thickness):
    return cv2.putText(image, text, position, font, font_size, color, thickness, cv2.LINE_AA)

def draw_ellipse(image, center, axes, angle, start_angle, end_angle, color, thickness):
    return cv2.ellipse(image, center, axes, angle, start_angle, end_angle, color, thickness)

def draw_polyline(image, points, is_closed, color):
    points = points.reshape((-1, 1, 2))
    return cv2.polylines(image, [points], is_closed, color)

# Main code
img = create_blank_image(512, 512)

img = draw_line(img, (0, 0), (255, 255), (147, 96, 44), 10)
img = draw_arrowed_line(img, (0, 255), (0, 255), (255, 0, 0), 10)
img = draw_rectangle(img, (384, 0), (510, 128), (0, 0, 255), 10)
img = draw_circle(img, (447, 63), 63, (0, 255, 0), -1)

font = cv2.FONT_HERSHEY_SIMPLEX
img = draw_text(img, 'mhamed', (10, 500), font, 4, (0, 255, 255), 10)

img = draw_ellipse(img, (256, 256), (100, 50), 0, 0, 180, 255, -1)

pts = np.array([[10, 5], [20, 30], [70, 20], [50, 10]], np.int32)
img = draw_polyline(img, pts, True, (0, 255, 255))

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
